# Traveling Salesman Problem - Genetic Algorithms

This notebook contains a pretty straightforward approach to solve the TSP, using a genetic algorithm.

### Some Julia Models

```julia
ζ = Symmetric(readdlm("cities.txt", Int64))
```

```julia
using DelimitedFiles

using Plots
using GraphRecipes

using Random.Random
using StatsBase
using Distributions
using LinearAlgebra
```

```julia
Random.shuffle(s::String) = isascii(s) ? s[randperm(end)] : join(shuffle!(collect(s)))
```

```julia
function Φ(
        p::String
    )::Float64
    
    ϕ = 0
    
    for i=1:length(p) - 1
        ϕ += ζ[Int(p[i]) - 64, Int(p[i + 1]) - 64]
    end
    
    return 1 / ϕ
end
```

```julia
function ∇(
        P::Array{String,1},
        ω::Array{Float64,1},
        μ::Int64
    )::Array{String,1}

    β = String[]
    
    for i=1:μ
        Σ = 0
        i = 0

        ϕ = rand(Uniform(0, 1))

        while Σ <= ϕ
            i = i % μ + 1
            Σ += ω[i]
        end

        push!(β, P[i])
    end
    
    return unique!(β)
end
```

```julia
function ⊗(
        β::Array{String,1},
        η::Int64,
        δ::String
    )::Array{String, 1} 
    
    βn = length(β)
    
    α = String[]
    
    for i=1:βn-1
        for j=i+1:βn
            c1 = Δ(β[i], β[j], δ, η)
            c2 = Δ(β[j], β[i], δ, η)
            append!(α, [c1, c2])
        end
    end
    
    return unique!(α)
end
```

```julia
function Δ(
        ρ1::String,
        ρ2::String,
        δ::String,
        η::Int64
    )::String
    
    θ = Char[]
    i = j = k = 1 

    while i <= η
    
        if δ[i] == '0'
            if ρ1[j] in θ
                j += 1
            else
                push!(θ, ρ1[j])
                i += 1
            end
                
        else
            if ρ2[k] in θ
                k += 1
            else
                push!(θ, ρ2[k])
                i += 1
            end
        end
    end
    
    return join(θ)
end
```

```julia
function ∏(
        P::Array{String,1},
        α::Array{String,1},
        μ::Int64
    )
    
    α = append!(P, α)
    
    Φv = [Φ(ι) for ι in α]
    k = sortperm(Φv, rev=true)

    α = α[k]
  
    return α[1 : min(μ, length(α))]
end
```

```julia
function γ(
        α::Array{String,1},
        η::Int64
    )
    
    γ1, γ2 = Random.rand(1:η, 2)
    
    l = length(α)
    
    if l >= 1
        λ = Random.rand(1:l)
    
    
        s = collect(α[λ])
    
        τ = s[γ1]
        s[γ1] = s[γ2]
        s[γ2] = τ
        
        α[λ] = join(s) 
    end
end
```

```julia
function Γ(
        η::Int64,
        μ::Int64,
        τ::Int64,
        δ::String
    )::Float64
    
    ψ = join([Char(64 + i) for i=1:η])
    P = [shuffle(ψ) for i=1:μ]
    
    for i=1:τ
        
        Φv = Float64[]
        for p in P
            push!(Φv, Φ(p))
        end
        
        ΦT = sum(Φv) 
        
        ω  = [ϕ / ΦT for ϕ in Φv]
                
        β = ∇(P, ω, μ)
        
        α = ⊗(β, η, δ)

        γ(α, η)
        
        P = ∏(P, α, μ)
    end
    
    Φavg = sum([Φ(ι) for ι in P]) / μ
    
    return Φavg 
end
```

```julia
x = collect(1:1:50)
y = Float64[]

for i in x
    push!(y, Γ(15, 500, i, "001100110011001"))
end

scatter(x, y, alpha=0.8)
```
